make a program that takes number of student and tell what grade did he got according to his percentage

```python
mark = []
tot = 0
print ("Enter Marks Obtained in 5 Subjects: ")
for m in range(5):
    mark.insert (m, input())
for m in range(5):
    tot = tot + int(mark[m])
avg = tot/5
if avg>=91 and avg <=100:
   print ("Your Grade is A1", "Excellent")
elif avg>=80 and avg <=90:
   print ("Your Grade is A-1", "VERY GOOD")
elif avg>=70 and avg <=79:
   print ("Your Grade is A", "GOOD")
elif avg>=60 and avg <=69:
   print ("Your Grade is B", "PASS")
elif avg>=50 and avg <=59:
   print ("Your Grade is C", "JUST PASS")
elif avg>=40 and avg <=49:
   print ("Your Grade is D", "FAIL: TRY IN NEXT EXAM")
```
